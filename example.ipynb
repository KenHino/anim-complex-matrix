```python
import numpy as np
from main import get_anim
from IPython.display import HTML
```

```python
time_steps = 20
size = 5
time = np.linspace(0, 20, time_steps)
data = np.zeros((time_steps, size, size), dtype=np.complex128)
for i in range(time_steps):
    magnitude = np.random.random((size, size)) + 0.5  # Random magnitudes > 0.5
    phase = time[i] + np.random.random((size, size)) * np.pi / 4
    data[i] = magnitude * np.exp(1j * phase)
np.save('time', time)
np.save('data', data)
```

```python
print(f"{data.shape=}, {data.dtype=}")
print(f"{time.shape=}, {time.dtype=}")
```
